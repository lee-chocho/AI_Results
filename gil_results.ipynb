**연구 정리**
* 실제 의료 데이터는 불균형 데이터가 많아 데이터를 그대로 사용하면 예측률이 낮아질 수 밖에 없다.
* 한편, 사용 가능한 데이터의 수는 오히려 부족한 경우가 많다.
* 따라서, 불균형 데이터로 인한 문제와 데이터 부족 문제를 해결하기 위해 oversampling 기법을 사용하였다.
* 실제 데이터는 Null 데이터가 많아 결측치를 채우기 위한 여러 방법들이 연구가 진행되고 있으나, XGboost 모델의 결측치를 무시하는 파라미터를 사용함으로써 결측치를 채우지 않고도 탐색적 연구가 가능하였다.
