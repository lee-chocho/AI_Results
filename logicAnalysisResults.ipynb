#**Overview**



*   타기업에서 연구한 내용을 내부적으로 응용해보기 위한 작업
*   있는 데이터를 최대로 잘 활용하기 위해 환자별 특성을 반영하여 feature뿐만 아니라 target도 결측값 처리를 함
*   많은 feature보다 중요한 feature를 잘 선택하여 모델에 적용하는 편이 좋은 성능을 보이는 것을 확인함
*   값 예측 알고리즘과 연결하지 않고 분류 모델 로직만 구현함

#**Results**

image.png

* AUROC Curve: x축이 FPR, y축이 TPR인 그래프에서 나타나는 곡선으로, 왼쪽 위가 뚱뚱할수록 좋은 성능을 보인다고 해석 가능
* Precision-Recall Curve: x축이 Recall, y축이 Precision인 그래프에서 나타나는 곡선으로, 오른쪽 위가 뚱뚱할수록 좋은 성능을 보인다고 해석 가능. +불균형한 데이터셋으로 만든 모델의 성능을 평가하는 방법 중 하나  
* Confusion Matrix: 테스트 데이터셋이 실제 예측한 클래스와 정답 클래스를 확인할 수 있는 매트릭스. 위의 Confusion Matrix에서는 불균형데이터셋으로 인해 class 1의 정답 데이터 수가 많이 집계된 것을 볼 수 있으나, 전반적으로 예측률이 나쁘지는 않아 보임.

image.png

* Accuracy: True를 True로 옳게 예측한 경우와 False를 False로 옳게 예측한 경우를 고려하는 지표. 단, domain의 bias를 고려해야함.
* Macro-average AUC: 다중 클래스 분류 모델의 성능을 측정하는 지표 중 하나인 AUC의 평균값을 나타내는 지표. (보통 AUC는 주로 이진분류에서 사용되지만 다중 분류에도 적용이 가능함. 각 클래스마다 AUC를 계산한 후 AUC값을 모두 더하여 클래스의 수로 나누어 평균값을 구함)
* F1 score: Precision과 Recall의 조화평균으로, 클래스가 불균형적일때 모델의 성능을 정확히 평가할 수 있음.
* Precision: True라고 분류한 것 중에서 실제로 True인게 얼마나 되는지 알고 싶을 경우 확인하는 지표
* Recall: 실제 True 중에서 True로 분류한게 얼마나 되는지 알고 싶을 경우 확인하는 지표
